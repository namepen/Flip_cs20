# Lecture 4. Eagar execution

텐서플로를 사용할때, 먼저 연산 과정을 그래프로 구성하는 Python 프로그램을 작성하게 된다. 프로그램은 그래프를 C++ runtime에 전달해주는 세션을 호출한다. 

계산과정을 명시해야 하므로 텐서플로를  'declarative programming(선언형 프로그래밍)'이라고 부른다.

텐서플로는 넓은 의미에서 머신 러닝 모델을 위한 프로그래밍 언어이자, 컴파일러라고 할 수 있다.

이러한 형식은 많은 장점을 가지고 있다.

하지만 몇 가지 단점도 가지고 있다.

단점을 극복하기 위해 Python에서 텐서플로를 바로 수행할 수 있는 방법이 바로 Eager execution 이다.

## Eager execution

GPU 가속과 automatic differentiation를 지원하는 Numpy와 비슷한 library이다. 머신러닝의 research와 experimentation를 좀 더 유연하게 만들어 준다.

```python
#Live Demo
import tensorflow as tf
import tensorflow.contrib.eager as tfe
tfe.enable_eager_execution()
```

```python
i = tf.constant(0)
while i <= 1000:
    i = tf.add(i, 1)
    print( 'I could do this all day! %d' % i)
```

eager execution을 사용하면 코드를 간결하게 만들어주고, Placeholder, session, lazy loading, scope 등등... 에 신경쓰지 않아도 된다.

```python
#Boilerplate
#x = tf.placeholder(tf.float32, shape=[1,1])
x = [[2.]]
m = tf.matmul(x, x)
''' before eager execution
print(m)

with tf.Session() as sess:
    m_out = sess.run(m, feed_dict={x: [[2.]]})
print(m_out)
'''

print(m)
```

### Use eager if you're

* a researcher and want a flexible framework
    - python control flow and data structures enable experimentation
    
    
* developing a new model
    - immediate error reporting simplifies debugging
    
    
* new to TensorFlow
    - eager execution lets you explore the TF API in the Python REPL
